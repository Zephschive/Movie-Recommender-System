```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
movies.dropna(inplace=True)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] =movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df =movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps= PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return "".join(y)  
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)
```

```python
similarity =cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import networkx as nx
import matplotlib.pyplot as plt


import pickle

new_df = pickle.load(open('movie_list.pkl', 'rb'))
similarity = pickle.load(open('similarity.pkl', 'rb'))


G = nx.Graph()


for index, row in new_df.iterrows():
    G.add_node(row['title'])


for i in range(len(similarity)):
    for j in range(i+1, len(similarity)):
        if similarity[i, j] > 0.5: 
            G.add_edge(new_df.iloc[i]['title'], new_df.iloc[j]['title'])

# Plot the graph
pos = nx.spring_layout(G)  # You can use different layout algorithms
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, font_size=8, node_size=1, font_color='black')
plt.title('Movie Recommender System Graph')
plt.show()
```

```python
import numpy as np
import pandas as pd
import pickle
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Load the preprocessed data and similarity matrix
new_df = pickle.load(open('movie_list.pkl', 'rb'))
similarity = pickle.load(open('similarity.pkl', 'rb'))

# Perform hierarchical clustering
linkage_matrix = linkage(similarity, "ward")

# Create a dendrogram
plt.figure(figsize=(15, 8))
dendrogram(linkage_matrix, labels=new_df['title'].values, leaf_rotation=90, leaf_font_size=8)
plt.title('Movie Recommender System Dendrogram')
plt.xlabel('Movies')
plt.ylabel('Distance')
plt.show()
```
